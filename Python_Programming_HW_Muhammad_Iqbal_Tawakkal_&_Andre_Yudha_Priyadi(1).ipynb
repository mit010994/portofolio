# SOAL 1

```python
list_pelat_mobil = ['B 1234 AB', 'B 6721 TY', 'B 1233 AY', 'B 6629 DD', 'B 1111 AM', 'B 6726 D', 'D 11223 KJ', 'AE 44677 GH', 'AE 67269 AA']

# Separate alphabet & number
numbers = []
for item in list_pelat_mobil:
    for subitem in item.split():
        if(subitem.isdigit()):
            numbers.append(subitem)
            
# Convert (number) string into integer
for i in range(0, len(numbers)):
    numbers[i] = int(numbers[i])

numbers
```

```python
# Generate value from the iterarion
check_pelat = []

for i in numbers:
    if i %2==0:
        check_pelat.append('Genap')
    else:
        check_pelat.append('Ganjil')

check_pelat
```

```python
# Count each value in list
counter1 = check_pelat.count('Genap')
counter2 = check_pelat.count('Ganjil')

print('Jumlah pelat bernomor Genap adalah:', counter1)
print('Jumlah pelat bernomor Ganjil adalah:', counter2)
```

# SOAL 4

```python
def string_processor (proses, kata):
  proses = proses.lower()
  vowel_a = []
  vowel_i = []
  vowel_u = []
  vowel_e = []
  vowel_o = []
  if proses == 'first_char':
    return str(kata)[0]
  elif proses == 'last_char':
    return str(kata)[-1]
  elif proses == 'num_chars':
    return len(kata)
  elif proses == 'num_vowels':
    for i in kata:
      i = i.lower()
      if i == 'a':
        vowel_a.append('1')
      elif i  == 'i':
        vowel_i.append('1')
      elif i  == 'u':
        vowel_u.append('1')
      elif i  == 'e':
        vowel_e.append('1')
      elif i  == 'o':
        vowel_o.append('1')
      num_vowels = len(vowel_a) + len(vowel_i) + len(vowel_u) + len(vowel_e) + len(vowel_o)
    return num_vowels
  else:
    print('invalid string processor')
```

```python
print(string_processor('first_char','merkaptan'))
print(string_processor('last_char','merkaptan'))
print(string_processor('num_chars','merkaptan'))
print(string_processor('num_vowels','merkaptan'))
```

```python
# Create new column with date conversion
df['Year_Order'] = df['Order Date'].dt.year 
df.head()
```

```python
# Calculate total order for each year in a specific region
def salesprofit_region(region, year = 2017):
    df_filter_region = df[(df.Year_Order == year)
                          & (df.Region == region)
                          ].reset_index()
    
    total_sales = df_filter_region.Sales.sum()
    total_profit = df_filter_region.Profit.sum()
    
    print('Region:',region,', year:',year,'- Total Sales: $',round(total_sales,2),', Total Profit: $',round(total_profit,2))
```

# SOAL 7

```python
import pandas as pd
import numpy as np

df = pd.read_excel('superstore.xls')
df
```

```python
# Create new column with date conversion
df['Year_Order'] = df['Order Date'].dt.year 
df.head()
```

```python
# Aggregate the sales if each customer
sales_user = df.groupby(['Customer ID','Customer Name','Order Year']).agg({'Sales':'sum'}).sort_values(('Customer ID'),ascending=True).reset_index()
sales_user
```

```python
# Merge sales performance dataframe with statistical feature
stat_desc = sales_user.groupby('Order Year')['Sales'].describe().reset_index()
df_grouped = sales_user.merge(stat_desc, on = 'Order Year')
df_grouped
```

```python
# Select only relevant columns
df_grouped = df_grouped[['Customer ID','Customer Name','Order Year','Sales','25%','50%','75%']]

# Iterate dataframe to create a segmentation
segment = []

for index, column in df_grouped.iterrows():
    if column['Sales'] >= column['50%']:
        segment.append('HIGH')
    elif column['Sales'] > column['25%'] and column['Sales'] < column['50%']:
        segment.append('MEDIUM')
    else:
        segment.append('LOW')

df_grouped['Segment'] = segment

df_grouped
```

```python
# Calculate total customers in each segment with aggregate function
user_segmented2017 = df_grouped[df_grouped['Order Year'] == 2017].groupby('Segment').agg({'Segment':'count'})

user_segmented2017 = user_segmented2017.rename(columns={'Segment': 'Total Costumers'})

user_segmented2017
```

```python
# Error handling
segment = []

try:
    for index, column in df_grouped.iterrows():
        if column['Sales'] >= column['50%']:
            segment.append('HIGH')
        elif column['Sales'] > column['25%'] and column['Sales'] < column['50%']:
            segment.append('MEDIUM')
        else:
            segment.append('LOW')

except:
    print('Error!')
```
