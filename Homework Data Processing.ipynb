## No.8

```python
revenue = pd.read_csv('01 telecom_revenue.csv')
```

```python
usage = pd.read_csv('02 telecom_usage.csv')
```

```python
revenue
```

```python
usage
```

```python
df_1 = revenue[revenue['Occupation'].isin(['Professional','Student','Crafts'])&(revenue['MonthlyRevenue']>10)]
```

```python
df_1.head()
```

```python
df_2 = usage[usage['CustomerCareCalls'].astype('str').str[0].isin(['2','4','6','8'])&usage['UnansweredCalls'] > usage['BlockedCalls']]
```

```python
df_2
```

```python
df_merged = df_1.merge(df_2,
                                    left_on = 'CustomerID',
                                    right_on = 'CustomerID',
                                    how = 'inner')
```

```python
df_merged
```

```python
df_merged.groupby('Occupation').agg({'DroppedCalls':'mean',
                                    'CustomerID':'count'})
```

```python
#Rata-rata dropped calls tertinggi terjadi pada pengguna dengan profesi Crafts, dengan jumlah CustomerID 1 untuk Crafts dan 7 untuk Professional
```
